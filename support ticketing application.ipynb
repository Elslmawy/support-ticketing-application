```python
import json
import uuid

# Load issue categories from external JSON file
def load_categories(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)['categories']

# Create a new ticket
def create_ticket(categories):
    print("Create a new support ticket")
    print("Select a category:")
    
    for idx, category in enumerate(categories.keys(), 1):
        print(f"{idx}. {category}")
    
    category_choice = int(input("Enter the number of the category: ")) - 1
    selected_category = list(categories.keys())[category_choice]
    
    print(f"Selected Category: {selected_category}")
    print("Select an issue type:")
    
    for idx, issue in enumerate(categories[selected_category], 1):
        print(f"{idx}. {issue}")
    
    issue_choice = int(input("Enter the number of the issue: ")) - 1
    selected_issue = categories[selected_category][issue_choice]
    
    ticket_id = uuid.uuid4()
    description = input("Describe your issue: ")
    
    ticket = {
        "id": str(ticket_id),
        "category": selected_category,
        "issue_type": selected_issue,
        "description": description
    }
    
    return ticket

# Main function to run the ticketing system
def main():
    categories = load_categories('categories.json')
    tickets = []
    
    while True:
        print("\n1. Create a new ticket\n2. View all tickets\n3. Exit")
        choice = input("Enter your choice: ")
        
        if choice == '1':
            ticket = create_ticket(categories)
            tickets.append(ticket)
            print(f"Ticket created successfully! Ticket ID: {ticket['id']}")
        elif choice == '2':
            if not tickets:
                print("No tickets found.")
            else:
                for ticket in tickets:
                    print(f"ID: {ticket['id']}, Category: {ticket['category']}, Issue: {ticket['issue_type']}, Description: {ticket['description']}")
        elif choice == '3':
            print("Exiting the system.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
